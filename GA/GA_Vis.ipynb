```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from itertools import chain

sns.set()
```

```python
df_final = pd.read_csv("csvs/final.csv")
df_final.rename({"Unnamed: 0":"id"}, axis=1, inplace=True)
df_final["pop"] = "1000"
df_final.head()
```

```python
df_0 = pd.read_csv("csvs/0.csv")
df_0.rename({"Unnamed: 0":"id"}, axis=1, inplace=True)
df_0["pop"] = "0"
df_0.head()
```

```python
df_500 = pd.read_csv("csvs/500.csv")
df_500.rename({"Unnamed: 0":"id"}, axis=1, inplace=True)
df_500["pop"] = "500"
df_500.head()
```

```python
df_all = pd.concat([df_final, df_500, df_0], ignore_index=True)
df_all.head()
```

```python
fig, axs = plt.subplots(2,2, figsize=(15, 15))

for y in range(len(axs)):
    for x in range(len(axs[0])):
        axs[y][x].set_xlabel("No of Tasks Completed")
        axs[y][x].set_ylabel("No of Unreachable Locs")

axs[0][0].set_title("All Generations")
axs[0][1].set_title("Generation 0")
axs[1][0].set_title("Generation 500")
axs[1][1].set_title("Generation 1000 (Final)")
    
sns.scatterplot(data=df_all, x="tasks_completed", y="no_unreachable_locs", hue="pop", ax=axs[0][0])
sns.scatterplot(data=df_0, x="tasks_completed", y="no_unreachable_locs", ax=axs[0][1], 
                color="green")
sns.scatterplot(data=df_500, x="tasks_completed", y="no_unreachable_locs", ax=axs[1][0], 
                color="darkorange")
sns.scatterplot(data=df_final, x="tasks_completed", y="no_unreachable_locs", ax=axs[1][1])

fig.savefig("unreachable_locs_tasks_completed.png")
```
