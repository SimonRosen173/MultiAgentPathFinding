```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

```python
df = pd.read_csv("benchmarks/no_agents_vs_t_tc.csv")
fig, axs = plt.subplots(2,2, figsize=(12,12))
sns.lineplot(x=df["no_agents"], y=df["time_taken_avg"], ax=axs[0][0])
sns.lineplot(x=df["no_agents"], y=df["time_taken_var"], ax=axs[0][1])
sns.lineplot(x=df["no_agents"], y=df["tasks_completed_avg"], ax=axs[1][0])
sns.lineplot(x=df["no_agents"], y=df["tasks_completed_var"], ax=axs[1][1])
```
